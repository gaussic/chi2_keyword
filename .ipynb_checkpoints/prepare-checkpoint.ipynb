```python
def segment(sentence, cut_all=False):
    sentences = re.sub('[\n\u3000。,，\?？\u00a0”“\(\)（）、\-—：/《》！～]', ' ', sentence)
    sentence = [x for x in jieba.cut(sentences, cut_all=cut_all) if x != ' ']
    return sentence
```

```python
document = MyDocuments(ROOT_DIR)
cnt = 0
from collections import Counter, defaultdict

counter = Counter()

cat_doc_cnt = defaultdict(int)
word_tot_cnt = Counter()
word_cat_cnt = defaultdict(Counter)


for cat, doc in document:
    doc = set(doc)
    cat_doc_cnt[cat] += 1
    word_tot_cnt.update(doc)
    word_cat_cnt[cat].update(doc)
        
    cnt += 1
    
    if cnt % 10000 == 0:
        print('Documents processed:', cnt, 'total words:', len(word_tot_cnt), ', time:', datetime.now().strftime('%H:%M:%S'))
```

```python
tot_doc = sum(cat_doc_cnt.values())
```

```python
def get_chi2(word):
    res = []
    for cat in os.listdir(ROOT_DIR):
        A = word_cat_cnt[cat][word]
        B = word_tot_cnt[word] - word_cat_cnt[cat][word]
        C = cat_doc_cnt[cat] - word_cat_cnt[cat][word]
        D = tot_doc - word_tot_cnt[word] - cat_doc_cnt[cat] + word_cat_cnt[cat][word]

        res.append((word, cat, (A*D - B*C)**2 / (A+B) / (C+D) / (A+C) / (B+D)))
    return res
    
```
