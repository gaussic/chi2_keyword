```python
def segment(sentence, cut_all=False):
    sentences = re.sub('[\n\u3000。,，\?？\u00a0”“\(\)（）、\-—：/《》！～]', ' ', sentence)
    sentence = [x for x in jieba.cut(sentences, cut_all=cut_all) if x != ' ']
    return sentence
```
